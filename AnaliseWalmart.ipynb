Vendas Semanais:
  Comparar vendas semanais entre diferentes lojas (se houver mais de uma).
  Analisar sazonalidade nas vendas (se houver).

```python
top_5_lojas = base_walmart.groupby('LOJA')['VENDA SEMANAL'].mean().sort_values(ascending=False).round(2)  # Calcula a média de vendas semanais para as lojas e ordena do maior para o menor
print(top_5_lojas.head(5), '\n')  # Exibe as 5 lojas com maiores médias de vendas semanais

top_5_less = base_walmart.groupby('LOJA')['VENDA SEMANAL'].mean().sort_values(ascending=True).round(2)  # Calcula a média de vendas semanais para as lojas e ordena do menor para o maior
print(top_5_less.head(5), '\n')  # Exibe as 5 lojas com menores médias de vendas semanais

desvio_padrao = top_5_lojas.std().round(2)  # Calcula o desvio padrão das médias de vendas semanais das lojas
print(f'Desvio Padrão de vendas semanais: {desvio_padrao}')  # Exibe o desvio padrão das médias de vendas semanais

l1 = base_walmart.loc[base_walmart['LOJA'] == 20]['VENDA SEMANAL'].mean().round(2)  # Calcula a média de vendas semanais para a loja 20
l2 = base_walmart.loc[base_walmart['LOJA'] == 4]['VENDA SEMANAL'].mean().round(2)  # Calcula a média de vendas semanais para a loja 4
l3 = base_walmart.loc[base_walmart['LOJA'] == 14]['VENDA SEMANAL'].mean().round(2)  # Calcula a média de vendas semanais para a loja 14
l4 = base_walmart.loc[base_walmart['LOJA'] == 13]['VENDA SEMANAL'].mean().round(2)  # Calcula a média de vendas semanais para a loja 13
l5 = base_walmart.loc[base_walmart['LOJA'] == 2]['VENDA SEMANAL'].mean().round(2)  # Calcula a média de vendas semanais para a loja 2

diferenca_1_2 = l1 - l2  # Calcula a diferença entre as médias de vendas semanais das lojas 1 e 2
diferenca_2_3 = l2 - l3  # Calcula a diferença entre as médias de vendas semanais das lojas 2 e 3
diferenca_3_4 = l3 - l4  # Calcula a diferença entre as médias de vendas semanais das lojas 3 e 4
diferenca_4_5 = l4 - l5  # Calcula a diferença entre as médias de vendas semanais das lojas 4 e 5

media_diferencas = ((diferenca_1_2 + diferenca_2_3 + diferenca_3_4 + diferenca_4_5) / 4).round(2)  # Calcula a média das diferenças entre as 5 lojas com maiores médias de vendas semanais
print(f'Média das diferenças entre as 5 lojas que mais rentabilizaram: {media_diferencas}')  # Exibe a média das diferenças
```

Feriados (Holiday_Flag):
  Avaliar o impacto dos feriados nas vendas.
  Comparar o desempenho de vendas em semanas com e sem feriados.
  Identificar quais feriados têm maior impacto nas vendas.

```python

correlacao_feriado_venda = base_walmart['FERIADO'].corr(base_walmart['VENDA SEMANAL'])  # Calcula a correlação entre os dias de feriado e as vendas semanais
print(f'Correlação entre dias de feriado e vendas da semana: {correlacao_feriado_venda:.3f}')  # Exibe a correlação entre feriados e vendas semanais

lojas_desejadas_top = [20, 4, 14, 13, 2]  # Lista das lojas desejadas (com maiores médias de vendas)

df_top5 = base_walmart[base_walmart['LOJA'].isin(lojas_desejadas_top)]  # Seleciona as linhas das 5 lojas desejadas
df_top5 = df_top5.sort_values(by='VENDA SEMANAL', ascending=False)  # Ordena as linhas pelo valor das vendas semanais de forma decrescente
display(df_top5.head(20))  # Exibe as primeiras 20 linhas do DataFrame resultante

lojas_desejadas_lower = [33, 44, 5, 36, 8]  # Lista das lojas desejadas (com menores médias de vendas)

df_lower5 = base_walmart[base_walmart['LOJA'].isin(lojas_desejadas_lower)]  # Seleciona as linhas das 5 lojas desejadas
display(df_lower5.head(20))  # Exibe as primeiras 20 linhas do DataFrame resultante

media_sem_feriados = base_walmart[base_walmart['FERIADO'] == 1]['VENDA SEMANAL'].mean().round(2)  # Calcula a média de vendas em dias de feriado
print(f'Média de vendas em dias que são feriados: {media_sem_feriados}')  # Exibe a média de vendas em dias de feriado

media_com_feriados = base_walmart[base_walmart['FERIADO'] == 0]['VENDA SEMANAL'].mean().round(2)  # Calcula a média de vendas em dias sem feriado
print(f'Média de vendas em dias que não são feriados: {media_com_feriados}')  # Exibe a média de vendas em dias sem feriado
```

Temperatura:
  Investigar a relação entre a temperatura e as vendas.
  Identificar padrões sazonais ou climáticos que afetam as vendas.

```python
correlacao_temperatura_vendas = base_walmart['TEMPERATURA'].corr(base_walmart['VENDA SEMANAL'])  # Calcula a correlação entre temperatura e vendas
print(f'Correlação entre temperatura e valor de vendas: {correlacao_temperatura_vendas:.5f}')  # Exibe a correlação entre temperatura e vendas

media_temperatura_geral = base_walmart['TEMPERATURA'].mean()  # Calcula a média da temperatura geral

top_20_vendas = base_walmart.sort_values(by='VENDA SEMANAL', ascending=False).head(20)  # Seleciona as 20 maiores vendas
lower_20_vendas = base_walmart.sort_values(by='VENDA SEMANAL', ascending=True).head(20)  # Seleciona as 20 menores vendas

media_temperatura_TOP20 = top_20_vendas['TEMPERATURA'].mean()  # Calcula a média da temperatura das 20 maiores vendas
media_temperatura_LOWER20 = lower_20_vendas['TEMPERATURA'].mean()  # Calcula a média da temperatura das 20 menores vendas

TempCelsius_geral = ((media_temperatura_geral - 32) / 1.8)  # Converte a média de temperatura geral para Celsius
TempCelsius_top = ((media_temperatura_TOP20 - 32) / 1.8)  # Converte a média de temperatura das 20 maiores vendas para Celsius
TempCelsius_lower = ((media_temperatura_LOWER20 - 32) / 1.8)  # Converte a média de temperatura das 20 menores vendas para Celsius

print(f'Média temperatura total: {TempCelsius_geral:.2f}°C\nMédia temperatura top 20 vendas: {TempCelsius_top:.2f}°C\nMédia temperatura piores 20 vendas: {TempCelsius_lower:.2f}°C')  # Exibe as médias de temperatura
```

Preço do Combustível (Fuel_Price):
  Analisar se há correlação entre o preço do combustível e as vendas.
  Verificar se mudanças nos preços do combustível afetam o comportamento de compra dos clientes.
  Analisar a relação entre a taxa de desemprego e as vendas.

```python
media_valor_combustivel = base_walmart['PRECO COMBUSTIVEL'].mean().round(2)  # Calcula a média do preço do combustível

top_10_vendas = base_walmart.sort_values(by='VENDA SEMANAL', ascending=False).head(10)  # Seleciona as 10 maiores vendas
media_top_10 = top_10_vendas['PRECO COMBUSTIVEL'].mean().round(2)  # Calcula a média do preço do combustível das 10 maiores vendas

low_10_vendas = base_walmart.sort_values(by='VENDA SEMANAL', ascending=True).head(10)  # Seleciona as 10 menores vendas
media_lower_10 = low_10_vendas['PRECO COMBUSTIVEL'].mean().round(2)  # Calcula a média do preço do combustível das 10 menores vendas

print(f'Média geral de combustível: {media_valor_combustivel} \nMédia combustível top 10 vendas: {media_top_10} \nMédia combustível 10 piores vendas: {media_lower_10}')  # Exibe as médias do preço do combustível

correlacao_combust_familia = base_walmart['PRECO COMBUSTIVEL'].corr(base_walmart['VAR MED DE PRECOS POR FAMILIA'])  # Calcula a correlação entre o preço do combustível e a média de consumo por família
print(f'Correlação entre preço do combustível e média de consumo por família: {correlacao_combust_familia:.5f}')  # Exibe a correlação entre preço do combustível e média de consumo por família

correlacao_desemprego_vendas = base_walmart['TX MD SEMANAL DESEMPREGO'].corr(base_walmart['VENDA SEMANAL'])  # Calcula a correlação entre a taxa de desemprego e as vendas semanais
print(f'Correlação entre tx de desemprego e venda semanal: {correlacao_desemprego_vendas:.5f}')  # Exibe a correlação entre a taxa de desemprego e as vendas semanais

TX_desemp_lower_10 = low_10_vendas['TX MD SEMANAL DESEMPREGO'].mean().round(2)  # Calcula a média da taxa de desemprego das 10 menores vendas
TX_desemp_top_10 = top_10_vendas['TX MD SEMANAL DESEMPREGO'].mean().round(2)  # Calcula a média da taxa de desemprego das 10 maiores vendas

print(f'Taxa média de desemprego top 10 semanas com mais vendas: {TX_desemp_top_10}\nTaxa média desemprego 10 piores semanas de venda: {TX_desemp_lower_10}')  # Exibe as médias da taxa de desemprego
```

Com base nas análises realizadas:

1. **Vendas Semanais:**
   - As lojas mais destacadas em termos de vendas semanais são as de números 20, 4, 14, 13 e 2, enquanto as menos destacadas são as de números 33, 44, 5, 36 e 38.
   - Sem informações geográficas específicas, é difícil determinar os fatores sociológicos por trás das variações nos valores médios de venda entre as lojas.

2. **Feriados:**
   - A correlação entre dias de feriado e vendas semanais é insignificante (0.037).
   - Embora não haja uma correlação direta entre feriados e aumento nas vendas, é interessante notar que as maiores arrecadações ocorreram em semanas comemorativas como Dia de Ação de Graças, Natal e Black Friday.
   - A sazonalidade de vendas é mais pronunciada no final do ano, especialmente em 2010 e 2011.
   - A média de vendas em dias de feriado é consideravelmente maior do que em dias normais.

3. **Temperatura:**
   - A correlação entre temperatura e vendas é insignificante (-0.06381).
   - Apesar disso, observa-se que as maiores vendas semanais ocorreram nos meses finais do ano.
   - As médias de temperatura para as 20 melhores e 20 piores semanas de vendas variam consideravelmente.

4. **Combustível e Taxa de Desemprego:**
   - Existe uma correlação fraca entre o preço do combustível e a média de consumo por família (-0.17064), indicando uma tendência inversa: quanto mais alto o preço do combustível, menor o consumo.
   - Similarmente, há uma correlação inversa entre a taxa de desemprego e as vendas semanais (-0.10618), sugerindo que um aumento na taxa de desemprego está associado a uma diminuição nas vendas.
   - As médias de preço do combustível e taxa de desemprego para as 10 melhores e 10 piores semanas de vendas mostram variações consistentes com essas correlações.

Em resumo, enquanto algumas tendências são observadas, como o aumento nas vendas durante feriados e a relação inversa entre preço do combustível/taxa de desemprego e vendas, outras variáveis, como temperatura, não parecem ter uma influência significativa nas vendas semanais.
